```python
home = '/scratch/pdenblaauwen/rootzone/rootzone/LS3MIP/evspsbl/Monthly_2001-2012/same_resolution/DJF'

CESM2 = xr.open_dataset(home + '/CESM2_land-hist_r1i1p1f1_DJF_2001-2012.nc' , engine = 'netcdf4')
CMCC_ESM2 = xr.open_dataset(home + '/CMCC-ESM2_land-hist_r1i1p1f1_DJF_2001-2012.nc'     , engine = 'netcdf4')
ESM_1_1 = xr.open_dataset(home + '/E3SM-1-1_land-hist_r1i1p11f1_DJF_2001-2012.nc' , engine = 'netcdf4')
EC_Earth3_veg = xr.open_dataset(home + '/EC-Earth3-Veg_land-hist_r1i1p1f1_DJF_2001-2012.nc' , engine = 'netcdf4')
HadGEM3_GC31_LL = xr.open_dataset(home + '/HadGEM3-GC31-LL_land-hist_r1i1p1f3_DJF_2001-2012.nc' , engine = 'netcdf4')
IPSL_CM6A_LR = xr.open_dataset(home + '/IPSL-CM6A-LR_land-hist_r1i1p1f1_DJF_2001-2012.nc' , engine = 'netcdf4')
MIROC6 = xr.open_dataset(home + '/MIROC6_land-hist_r1i1p1f1_DJF_2001-2012.nc' , engine = 'netcdf4')
UKESM1_0_LL = xr.open_dataset(home + '/UKESM1-0-LL_land-hist_r1i1p1f2_DJF_2001-2012.nc' , engine = 'netcdf4')
DOLCE_v3_monthly_sg = xr.open_dataset(home + '/DOLCE_v3_hfls_2001-2012_DJF_Same_grid.nc')
FBP_monthly_sg = xr.open_dataset(home + '/FLUXCOM_BESS_PML_monthly_2001-2012_DJF_Same_grid.nc')
Average = xr.open_dataset(home + '/LS3MIP_avg_DJF_2001-2012.nc' , engine = 'netcdf4')
Mean = xr.open_dataset(home + '/LS3MIP_mean_DJF_2001-2012.nc' , engine = 'netcdf4')

LM1_monthly = CESM2
LM2_monthly = CMCC_ESM2
LM3_monthly = ESM_1_1
LM4_monthly = EC_Earth3_veg
LM5_monthly = HadGEM3_GC31_LL
LM6_monthly = IPSL_CM6A_LR
LM7_monthly = MIROC6
LM8_monthly = UKESM1_0_LL
LS3MIP_average = Average
LS3MIP_mean = Mean
DOLCE_v3 = DOLCE_v3_monthly_sg
FBP = FBP_monthly_sg

label = ['C2' , 'CE' , 'E3' , 'EE' , 'HG' ,'IP' ,'M6' ,'UK' , 'LSM mean' , 'DO', 'FB']
DATA = [LM1_monthly , LM2_monthly , LM3_monthly , LM4_monthly , LM5_monthly ,LM6_monthly ,LM7_monthly ,LM8_monthly , Average , DOLCE_v3 , FBP]
number_d = np.arange(1 , 11 , 1)
```

```python
"""Parameters"""

L_60 = 30
L_eq = 90

"""Converting to mm/d"""
t_day = 86400 #seconds
factor = t_day

#LM1

"""Removing Antarctica"""

LM1_monthly_weighted_Ant = LM1_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM1_monthly_weighted_SH  = LM1_monthly.evspsbl[: , L_60:L_eq , :] * factor

#LM2


"""Removing Antarctica"""

LM2_monthly_weighted_Ant = LM2_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM2_monthly_weighted_SH  = LM2_monthly.evspsbl[: , L_60:L_eq , :] * factor

#LM3

"""Removing Antarctica"""

LM3_monthly_weighted_Ant = LM3_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM3_monthly_weighted_SH  = LM3_monthly.evspsbl[: , L_60:L_eq , :] * factor

#LM4

"""Removing Antarctica"""

LM4_monthly_weighted_Ant = LM4_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM4_monthly_weighted_SH  = LM4_monthly.evspsbl[: , L_60:L_eq , :] * factor

#LM5

"""Removing Antarctica"""

LM5_monthly_weighted_Ant = LM5_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM5_monthly_weighted_SH  = LM5_monthly.evspsbl[: , L_60:L_eq , :] * factor

#LM6

"""Removing Antarctica"""

LM6_monthly_weighted_Ant = LM6_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM6_monthly_weighted_SH  = LM6_monthly.evspsbl[: , L_60:L_eq , :] * factor

#LM7

"""Removing Antarctica"""

LM7_monthly_weighted_Ant = LM7_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM7_monthly_weighted_SH  = LM7_monthly.evspsbl[: , L_60:L_eq , :] * factor

#LM8

"""Removing Antarctica"""

LM8_monthly_weighted_Ant = LM8_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM8_monthly_weighted_SH  = LM8_monthly.evspsbl[: , L_60:L_eq , :] * factor

#DOLCE v3

labda = 2.45 * 10**6 #MJ kg^-1
rho_w = 1000  #kg m ^-3
t_day = 86400 #seconds
factor_d = (t_day * 1000)/(rho_w * labda)

"""Removing Antarctica"""

DOLCE_v3_monthly_weighted_Ant = DOLCE_v3_monthly_sg.hfls[: , L_60: , :] * factor_d

"""-60 until 0 degrees"""

DOLCE_v3_monthly_weighted_SH  = DOLCE_v3_monthly_sg.hfls[: , L_60:L_eq , :] * factor_d

#FBP

"""Removing Antarctica"""

FBP_monthly_weighted_Ant = FBP_monthly_sg.Evaporation[: , L_60: , :]

"""-60 until 0 degrees"""

FBP_monthly_weighted_SH  = FBP_monthly_sg.Evaporation[: , L_60:L_eq , :]

#Average

"""Removing Antarctica"""

Average_monthly_weighted_Ant = LS3MIP_average.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

Average_monthly_weighted_SH  = LS3MIP_average.evspsbl[: , L_60:L_eq , :] * factor

#Mean

"""Removing Antarctica"""

Mean_monthly_weighted_Ant = LS3MIP_mean.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

Mean_monthly_weighted_SH  = LS3MIP_mean.evspsbl[: , L_60:L_eq , :] * factor
```
