```python
data = {'DOLCE_V2' : DOLCE_V2_1_20032009 , 'DOLCE_V3' : DOLCE_V3_20032009 , 'CLASS':CLASS_20032009 , 'LS3MIP_average':LS3MIP_average_20032009}
data = pd.DataFrame(data = data)

print('Some statistics of the monthly datasets')
print('-----' * 16)
display(data.describe())

data1 = [DOLCE_V2_1_20032009 , DOLCE_V3_20032009 , CLASS_20032009] 
data2 = [LS3MIP_average_20032009]
 
fig = plt.figure(figsize =(10, 4))
print('-----' * 16)
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(data1, patch_artist = True , boxprops = dict(facecolor = 'lightblue') , positions = [2 , 3 ,  4] , widths = 0.25)
bp = ax.boxplot(data2, patch_artist = True , boxprops = dict(facecolor = 'lightgreen') , positions = [1 ] , widths = 0.25)
ax.set_xticks([1 , 2 , 3 , 4 ] , ['LS3MIP_average' , 'DOLCE_V2_1' , 'DOLCE_V3' , 'CLASS'])
ax.set_ylabel('Mean daily evaporation [mm/d]')
ax.set_title('Mean global daily evapotranspiration monthly average between the years 2003 and 2009')
plt.axvline(1.5 , color = 'k' , linestyle = 'dashed')

# show plot
plt.grid()
plt.show()
```
