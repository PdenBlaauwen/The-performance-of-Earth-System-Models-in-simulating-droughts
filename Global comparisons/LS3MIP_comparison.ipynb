```python
"""Parameters"""

L_60 = 30
L_eq = 90

"""Converting to mm/d"""
t_day = 86400 #seconds
factor = t_day

#LM1

"""Removing Antarctica"""

LM1_monthly_weighted_Ant = LM1_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM1_monthly_weighted_SH  = LM1_monthly.evspsbl[: , L_60:L_eq , :] * factor

#LM2


"""Removing Antarctica"""

LM2_monthly_weighted_Ant = LM2_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM2_monthly_weighted_SH  = LM2_monthly.evspsbl[: , L_60:L_eq , :] * factor

#LM3

"""Removing Antarctica"""

LM3_monthly_weighted_Ant = LM3_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM3_monthly_weighted_SH  = LM3_monthly.evspsbl[: , L_60:L_eq , :] * factor

#LM4

"""Removing Antarctica"""

LM4_monthly_weighted_Ant = LM4_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM4_monthly_weighted_SH  = LM4_monthly.evspsbl[: , L_60:L_eq , :] * factor

#LM5

"""Removing Antarctica"""

LM5_monthly_weighted_Ant = LM5_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM5_monthly_weighted_SH  = LM5_monthly.evspsbl[: , L_60:L_eq , :] * factor

#LM6

"""Removing Antarctica"""

LM6_monthly_weighted_Ant = LM6_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM6_monthly_weighted_SH  = LM6_monthly.evspsbl[: , L_60:L_eq , :] * factor

#LM7

"""Removing Antarctica"""

LM7_monthly_weighted_Ant = LM7_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM7_monthly_weighted_SH  = LM7_monthly.evspsbl[: , L_60:L_eq , :] * factor

#LM8

"""Removing Antarctica"""

LM8_monthly_weighted_Ant = LM8_monthly.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

LM8_monthly_weighted_SH  = LM8_monthly.evspsbl[: , L_60:L_eq , :] * factor

#DOLCE v3

labda = 2.45 * 10**6 #MJ kg^-1
rho_w = 1000  #kg m ^-3
t_day = 86400 #seconds
factor_d = (t_day * 1000)/(rho_w * labda)

"""Removing Antarctica"""

DOLCE_v3_monthly_weighted_Ant = DOLCE_v3_monthly_sg.hfls[: , L_60: , :] * factor_d

"""-60 until 0 degrees"""

DOLCE_v3_monthly_weighted_SH  = DOLCE_v3_monthly_sg.hfls[: , L_60:L_eq , :] * factor_d

#FBP

"""Removing Antarctica"""

FBP_monthly_weighted_Ant = FBP_monthly_sg.Evaporation[: , L_60: , :]

"""-60 until 0 degrees"""

FBP_monthly_weighted_SH  = FBP_monthly_sg.Evaporation[: , L_60:L_eq , :]

#Average

"""Removing Antarctica"""

Average_monthly_weighted_Ant = LS3MIP_average.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

Average_monthly_weighted_SH  = LS3MIP_average.evspsbl[: , L_60:L_eq , :] * factor

#Mean

"""Removing Antarctica"""

Mean_monthly_weighted_Ant = LS3MIP_mean.evspsbl[: , L_60: , :] * factor

"""-60 until 0 degrees"""

Mean_monthly_weighted_SH  = LS3MIP_mean.evspsbl[: , L_60:L_eq , :] * factor
```
