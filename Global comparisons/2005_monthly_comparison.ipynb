```python
plt.figure(figsize = (14 ,6 ))
plt.title('Mean global daily evapotranspiration monthly average for the year 2005')

plt.xticks(np.arange(0 , 12 , 1) , ['Jan' , 'Feb' , 'Mar' , 'Apr' , 'May' , 'Jun' , 'Jul' , 'Aug' , 'Sep' , 'Oct' , 'Nov' , 'Dec'])
plt.grid()
plt.xlabel('Month [-]')
plt.ylabel('Evapotranspiration [mm/d]')
plt.plot(np.arange(0 , 12 , 1) , LFE_2005 , 'C1o--' , label = 'LandFluxEVAL' , alpha = 0.4)
plt.plot(np.arange(0 , 12 , 1) , DOLCE_V2_1_2005 , 'C2o--' , label = 'DOLCE_V2_1' , alpha = 0.4)
plt.plot(np.arange(0 , 12 , 1) , DOLCE_V3_2005 , 'C3o--' , label = 'DOLCE_V3' , alpha = 0.4)
plt.plot(np.arange(0 , 12 , 1) , CLASS_2005 , 'C4o--' , label = 'CLASS' , alpha = 0.4)
plt.plot(np.arange(0 , 12 , 1) , LS3MIP_average_2005 , 'kd-' , label = 'LS3MIP_average' , alpha = 0.8)

plt.legend();
```

```python
data = {'LFE' : LFE_2005 , 'DOLCE_V2' : DOLCE_V2_1_2005 , 'DOLCE_V3' : DOLCE_V3_2005 , 'CLASS':CLASS_2005 , 'LS3MIP':LS3MIP_average_2005}
data = pd.DataFrame(data = data)

print('Some statistics of the monthly datasets')
print('-----' * 16)
display(data.describe())

data1 = [LFE_2005 , DOLCE_V2_1_2005 , DOLCE_V3_2005 , CLASS_2005] 
data2 = [LS3MIP_average_2005]
 
fig = plt.figure(figsize =(10, 4))
print('-----' * 16)
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(data1, patch_artist = True , boxprops = dict(facecolor = 'lightblue') , positions = [2 , 3 ,  4 , 5] , widths = 0.25)
bp = ax.boxplot(data2, patch_artist = True , boxprops = dict(facecolor = 'lightgreen') , positions = [1 ] , widths = 0.25)
ax.set_xticks([1 , 2 , 3 , 4 , 5] , ['LS3MIP_average' ,   'LandFlux_EVAL' , 'DOLCE_V2_1' , 'DOLCE_V3' , 'CLASS'])
ax.set_ylabel('Mean daily evaporation [mm/d]')
ax.set_title('Mean global daily evapotranspiration monthly average for the year 2005')
plt.axvline(1.5 , color = 'k' , linestyle = 'dashed')

# show plot
plt.grid()
plt.show()
```
