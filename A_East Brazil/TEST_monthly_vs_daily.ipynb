```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
```

```python
P = np.random.randint(0 , 10 , 150)
ET = np.random.randint(0 , 5 , 150)
S = np.zeros(150)

for i in range(len(P)):
    S[i] = S[i - 1] + P[i] - ET[i]
    
S2 = np.zeros(5)

for i in range(len(S2)):
    S2[i] = S2[i - 1] + (np.mean(P[30 * i : 30 * i + 30]) - np.mean(ET[30 * i : 30 * i + 30])) * 30
```

```python
plt.plot([29 , 59 , 89 , 119 , 149] , S2 , 'ro' , label = 'Monthly mean data used')
plt.plot(S , label = 'Daily data used')
plt.grid()
plt.xlabel('Number of days [-]')
plt.ylabel('Storage [mm]')
plt.legend();
```
