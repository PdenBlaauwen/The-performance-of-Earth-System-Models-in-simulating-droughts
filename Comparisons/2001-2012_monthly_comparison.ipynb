```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
"""2001-2012"""

DOLCE_V2_1_20012012 = np.array([0.70819159, 0.75005787, 0.83842034, 0.93954201, 1.05791243,
       1.28523161, 1.31291568, 1.18362507, 1.01860033, 0.86451922,
       0.77029181, 0.69252451])

DOLCE_V3_20012012 = np.array([0.79005776, 0.85329776, 0.94376438, 1.07878903, 1.1969472 ,
       1.37341292, 1.41034155, 1.27673229, 1.09624695, 0.94356362,
       0.8506038 , 0.77715528])

LS3MIP_average_20012012 = np.array([0.74359322, 0.78210487, 0.83621516, 0.94845062, 1.10521865,
       1.26213968, 1.30724617, 1.18319185, 1.01644916, 0.87016771,
       0.78118036, 0.73560587])
```

```python
"""Plotting"""

plt.figure(figsize = (14 ,8 ))
plt.title('Mean global daily evapotranspiration monthly average for the years 2001-2012')

plt.xticks(np.arange(0 , 12 , 1) , ['Jan' , 'Feb' , 'Mar' , 'Apr' , 'May' , 'Jun' , 'Jul' , 'Aug' , 'Sep' , 'Oct' , 'Nov' , 'Dec'])
plt.grid()
plt.xlabel('Month [-]')
plt.ylabel('Evapotranspiration [mm/d]')
plt.plot(np.arange(0 , 12 , 1) , DOLCE_V2_1_20012012 , 'C1o--' , label = 'DOLCE_V2_1' , alpha = 0.4)
plt.plot(np.arange(0 , 12 , 1) , DOLCE_V3_20012012 , 'C2o--' , label = 'DOLCE_V3' , alpha = 0.4)
plt.plot(np.arange(0 , 12 , 1) , LS3MIP_average_20012012 , 'kd-' , label = 'LS3MIP_average' , alpha = 0.8)

plt.legend();
```

```python
"""Boxplots"""

data = {'DOLCE_V2' : DOLCE_V2_1_20012012 , 'DOLCE_V3' : DOLCE_V3_20012012  , 'LS3MIP_average':LS3MIP_average_20012012}
data = pd.DataFrame(data = data)

print('Some statistics of the monthly average')
print('-----' * 13)
display(data.describe())

data1 = [DOLCE_V2_1_20012012 , DOLCE_V3_20012012] 
data2 = [LS3MIP_average_20012012]
 
fig = plt.figure(figsize =(10, 4))
print('-----' * 13)
print()
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(data1, patch_artist = True , boxprops = dict(facecolor = 'lightblue') , positions = [2 , 3] , widths = 0.2)
bp = ax.boxplot(data2, patch_artist = True , boxprops = dict(facecolor = 'lightgreen') , positions = [1] , widths = 0.2)
ax.set_xticks([1 , 2 , 3] , ['LS3MIP_average' , 'DOLCE_V2_1' , 'DOLCE_V3'])
ax.set_ylabel('Mean daily evaporation [mm/d]')
ax.set_title('Mean global daily evapotranspiration monthly average between 2001 and 2012')
plt.axvline(1.5 , color = 'k' , linestyle = 'dashed')

# show plot
plt.grid()
plt.show()
```
